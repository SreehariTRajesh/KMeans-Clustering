```python
from sklearn.cluster import KMeans
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
```

```python
df=pd.read_csv('/home/rajesh/Downloads/abalone.csv')
df.head()
```

```python
km=KMeans(n_clusters=3)
km
```

```python
y_predicted=km.fit_predict(df[['LongestShell','Diameter','Height','WholeWeight','ShuckedWeight','VisceraWeight','ShellWeight','Rings']])
y_predicted
```

```python
df['class']=y_predicted
df.head()
```

```python
df1=df[df['class']==0]
df2=df[df['class']==1]
df3=df[df['class']==2]
plt.scatter(df1['LongestShell'],df1['Diameter'],marker='.',color='blue')
plt.scatter(df2['LongestShell'],df2['Diameter'],marker='.',color='green')
plt.scatter(df3['LongestShell'],df3['Diameter'],marker='.',color='red')
plt.legend()
plt.show()
```

```python
import numpy as np
np.array(df[['LongestShell']])
```

```python
scaler = MinMaxScaler()
scaler.fit(df[['LongestShell']])
df['LongestShell']=scaler.transform(df[['LongestShell']])
df
```

```python
scaler.fit(df[['Diameter']])
df['Diameter']=scaler.transform(df[['Diameter']])
df
```

```python
scaler.fit(df[['Height']])
df['Height']=scaler.transform(df[['Height']])
scaler.fit(df[['WholeWeight']])
df['WholeWeight']=scaler.transform(df[['WholeWeight']])
scaler.fit(df[['ShuckedWeight']])
df['ShuckedWeight']=scaler.transform(df[['ShuckedWeight']])
scaler.fit(df[['VisceraWeight']])
df['VisceraWeight']=scaler.transform(df[['VisceraWeight']])
scaler.fit(df[['ShellWeight']])
df['ShellWeight']=scaler.transform(df[['ShellWeight']])
scaler.fit(df[['Rings']])
df['Rings']=scaler.transform(df[['Rings']])
df
```

```python
km=KMeans(n_clusters=3)
predicted=km.fit_predict(df[['LongestShell','Diameter','Height','WholeWeight','ShuckedWeight','VisceraWeight','ShellWeight','Rings']])
predicted
```

```python
df['cluster']=predicted
df0=df[df['cluster']==0]
df1=df[df['cluster']==1]
df2=df[df['cluster']==2]
plt.scatter(df0['Height'],df0['Diameter'],marker='+',color='green')
plt.scatter(df1['Height'],df1['Diameter'],marker='+',color='blue')
plt.scatter(df2['Height'],df2['Diameter'],marker='+',color='red')
plt.xlabel('Height')
plt.ylabel('Diameter')
plt.show()
```

```python
plt.scatter(df0['WholeWeight'],df0['Diameter']+0.2,marker='.',color='green')
plt.scatter(df1['WholeWeight'],df1['Diameter']+0.1,marker='.',color='blue')
plt.scatter(df2['WholeWeight'],df2['Diameter'],marker='.',color='red')
plt.xlabel('WholeWeight')
plt.ylabel('Diameter')
plt.show()
```

```python
label=['LongestShell','Diameter','Height','WholeWeight','ShuckedWeight','VisceraWeight','ShellWeight','Rings']
for i in label:
    for j in label:
        if i !=j:
            plt.scatter(df0[i],df0[j],marker='+',color='green')
            plt.scatter(df1[i],df1[j],marker='+',color='blue')
            plt.scatter(df2[i],df2[j],marker='+',color='red')
            plt.xlabel(i)
            plt.ylabel(j)
            plt.show()
            
```
