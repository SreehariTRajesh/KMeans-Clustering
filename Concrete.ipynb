```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('/home/rajesh/Downloads/concrete.csv')
df.head()
```

```python
plt.scatter(df.Water,df.CoarseAggregate,marker='.',color='green')
```

```python
cols=df.columns
```

```python
cols
```

```python
scaler=MinMaxScaler()
for col in cols:
    scaler.fit(df[[col]])
    df[col]=scaler.transform(df[[col]])
    
```

```python
df.head()
```

```python
km = KMeans()
y=km.fit_predict(df[['Cement','BlastFurnaceSlag','FlyAsh','Water','Superplasticizer','CoarseAggregate','FineAggregate','Age','CompressiveStrength']])
```

```python
print(y)
df['cluster']=y
```

```python
df0=df[df['cluster']==0]
df1=df[df['cluster']==1]
df2=df[df['cluster']==2]
df3=df[df['cluster']==3]
df4=df[df['cluster']==4]
df5=df[df['cluster']==5]
df6=df[df['cluster']==6]
df7=df[df['cluster']==7]
```

```python
for col1 in cols:
    for col2 in cols:
        if col1!=col2:
            plt.scatter(df0[col1],df0[col2],marker='.',color='green')
            plt.scatter(df1[col1],df1[col2],marker='.',color='red')
            plt.scatter(df2[col1],df2[col2],marker='.',color='blue')
            plt.scatter(df3[col1],df3[col2],marker='.',color='yellow')
            plt.scatter(df4[col1],df4[col2],marker='.',color='violet')
            plt.scatter(df5[col1],df5[col2],marker='.',color='purple')
            plt.scatter(df6[col1],df6[col2],marker='.',color='orange')
            plt.scatter(df7[col1],df7[col2],marker='.',color='grey')
            plt.xlabel(f'{col1}')
            plt.ylabel(f'{col2}')
            plt.show()
```
